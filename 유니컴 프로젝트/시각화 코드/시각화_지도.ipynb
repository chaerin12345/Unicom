```python
import csv
import pandas as pd
from geopy.geocoders import Nominatim

input_file = r'/Users/crlee/융복합프로젝트/기업120_설립년도_규모_근로자수_위치_복지.csv'
output_file = r'기업위도경도_추출.csv'

geolocator = Nominatim(user_agent="my_geocoder", timeout=10)

# Read the input CSV file using pandas
df = pd.read_csv(input_file, encoding='utf-8-sig')

# Open the output CSV file for writing
with open(output_file, "w", newline="", encoding='utf-8-sig') as outfile:
    writer = csv.writer(outfile)

    # Write the header row with added "위도(latitude)" and "경도(longitude)" columns
    writer.writerow(df.columns.tolist() + ["위도(latitude)", "경도(longitude)"])

    # Loop through each row in the DataFrame
    for index, row in df.iterrows():
        address = row['위치']
        location = geolocator.geocode(address)

        if location:
            latitude = location.latitude
            longitude = location.longitude
        else:
            latitude = ""
            longitude = ""

        # Write the original row data along with latitude and longitude to the output CSV file
        writer.writerow(row.tolist() + [latitude, longitude])

print("위도 경도 추출이 완료되었습니다.")
```

```python
import pandas as pd
from geopy.geocoders import Nominatim

df = pd.read_csv('/Users/crlee/융복합프로젝트/기업140_설립년도_규모_근로자수_위치_복지.csv', encoding='euc-kr')
df

df2 = df['위치']

def get_lat_long_from_address(address):
    geolocator = Nominatim(user_agent="myGeocoder")
    location = geolocator.geocode(address)
    if location:
        return location.latitude, location.longitude
    else:
        return None, None


# 주소 리스트로부터 위도와 경도 추출
data = {'위치': df2, '위도': [], '경도': []}
for address in df2:
    latitude, longitude = get_lat_long_from_address(address)
    data['위도'].append(latitude)
    data['경도'].append(longitude)

# 데이터프레임 생성
df_result = pd.DataFrame(data)

# 결과 출력
print(df_result)
df_result.to_csv('위치위도경도.csv',  index=False, encoding='utf-8-sig')
```

```python
import folium
import json

# 전국 지도 경계 GeoJSON 파일 경로
korea_geojson_file = '/path/to/korea.geojson'

# 전국 지리적 경계 정보를 로드
with open(korea_geojson_file, 'r', encoding='utf-8') as f:
    korea_geojson = json.load(f)

# 서울시 지리적 경계 정보 추출
seoul_geojson = {
    'type': 'FeatureCollection',
    'features': [feature for feature in korea_geojson['features'] if feature['properties']['name'] == '서울특별시']
}

# 서울시를 칠할 색상
seoul_color = 'blue'

# 서울시 지도 초기 위치 설정
seoul_map = folium.Map(location=[37.566, 126.978], zoom_start=11)

# GeoJSON으로 서울시 지리적 경계를 칠하고 맵에 추가
folium.GeoJson(seoul_geojson,
               style_function=lambda feature: {
                   'fillColor': seoul_color,
                   'color': 'black',
                   'weight': 2,
                   'fillOpacity': 0.5,
               }).add_to(seoul_map)

# 맵을 HTML 파일로 저장
seoul_map.save('seoul_map.html')
```
