### corp 기업 table 생성

```python
# 남녀성비 계산
df['사원수'] = df['남사원수'] + df['여사원수']
df['남사원수성비']= df['남사원수']/df['사원수']
df['여사원수성비']= df['여사원수']/df['사원수']
```

```python
# 3년 평균 매출
df2['기업매출3년평균'] = df2.loc[:, ['2020_매출액', '2021_매출액', '2022_매출액']].mean(axis=1)
```

```python
# 3년 평균 영업이익
df5['영업이익3년평균'] = df5.loc[:, ['2020영업이익', '2021영업이익', '2022영업이익']].mean(axis=1)
```

```python
# 2022 연봉
df4[2022]
```

```python
# 기업명을 기준으로 merge
```

```python
from functools import reduce

dfs = [df, df2, df3, df4, df5]
merged_df = reduce(lambda left, right: pd.merge(left, right, on='기업명'), dfs)
```

```python
# 원하는 행만 뽑기 
df_corp = pd.DataFrame(columns=['corp_code','corp_name','corp_m','corp_w','corp_sales','corp_sm','corp_aug','corp_est','corp_sal','corp_add','corp_long','corp_lati','corp_wel','corp_ind','corp_ope'])
```

```python

i = 0
while i < len(merged_df):
    df_corp.loc[i] = [merged_df['기업코드'][i],merged_df['기업명'][i],merged_df['남사원수성비'][i],merged_df['여사원수성비'][i],merged_df['기업매출3년평균'][i],merged_df['사원수'][i],merged_df['평균근속연수'][i],merged_df['설립년도'][i],merged_df[2022][i],merged_df['위치'][i],merged_df['경도'][i],merged_df['위도'][i],merged_df['복지'][i],merged_df['산업군'][i],merged_df['영업이익3년평균'][i]]
    i += 1
```

```python
# corp 저장
```

```python
df_corp.to_csv("corp.csv",index=False, encoding='cp949')
```
