### 기업명 설립년도 위치 복지 크롤링 

```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.select import Select
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
import time
import pyperclip
import requests
from bs4 import BeautifulSoup
import pandas as pd


### 빈 기업 df 만들기
df = pd.DataFrame(columns=['기업명','설립년도','기업규모','근로자수','위치','복지'])
com = pd.read_csv('기업_140_산업군.csv')

i = 0
while i < len(com):
    driver = webdriver.Chrome()
    url = 'https://www.work.go.kr/seekWantedMain.do'
    driver.get(url)
    
    company = com['기업명'][i]
    time.sleep(3)
    search_input = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, 'topQuery')))
    search_input.click()

    # 검색 입력
    search_input.send_keys(company)

    # 검색 실행
    search_input.send_keys(Keys.ENTER)

    # 검색 결과 대기 시간 (예: 3초)
    time.sleep(3)

    # 기업정보 버튼 클릭
    # 클래스로 버튼 요소 찾기
    try:
        # 기업정보 버튼 찾기
        company_info_button = driver.find_element(by="xpath", value="//a[text()='기업정보']")

        # 버튼 클릭
        company_info_button.click()

        # 기업정보 새창으로 전환
        driver.switch_to.window(driver.window_handles[2])
        
        # 설립년도
        year_id = '//*[@id="divContentsPrint"]/div/div/div[3]/ul/li[1]/p'
        year = driver.find_element(By.XPATH, year_id)
        # 기업규모
        size_id ='//*[@id="divContentsPrint"]/div/div/div[3]/ul/li[2]/p'
        size = driver.find_element(By.XPATH, size_id)
        # 근로자수
        employee_id = '//*[@id="divContentsPrint"]/div/div/div[3]/ul/li[4]/p'
        employee = driver.find_element(By.XPATH, employee_id)


        # 페이지 소스 가져오기
        page_source = driver.page_source

        # BeautifulSoup으로 파싱
        soup = BeautifulSoup(page_source, 'html.parser')

        # ID로 요소 찾기
        element = soup.find(id="address1")

        # 텍스트 추출
        address = element["value"]


        # 클래스로 요소 찾기 및 텍스트 추출
        elements = soup.find_all(class_="wel03")
        text1 = [element.get_text(strip=True) for element in elements]

        elements = soup.find_all(class_="wel04 l_bd")
        text2 = [element.get_text(strip=True) for element in elements]

        elements = soup.find_all(class_="wel05")
        text3 = [element.get_text(strip=True) for element in elements]

        elements = soup.find_all(class_="wel06 l_bd")
        text4 = [element.get_text(strip=True) for element in elements]

        # 복지 개수 세기 0 ~ 4
        text_list = [text1, text2, text3, text4]
        value = sum([1 for text in text_list if text])

        # 데이터프레임에 추가
        df.loc[i] = [company, year.text,size.text,employee.text, address,value]
        
    except NoSuchElementException:
        try:
            # 연관 검색어 맨 처음 요소 선택
            related_search_button = driver.find_element(by="xpath", value="//*[@id='recommendDiv']/a/em")
            related_search_button.click()

            # 기업정보 버튼 찾기
            company_info_button = driver.find_element(by="xpath", value="//a[text()='기업정보']")

            # 버튼 클릭
            company_info_button.click()

            # 기업정보 새창으로 전환
            driver.switch_to.window(driver.window_handles[1])
            
            # 설립년도
            year_id = '//*[@id="divContentsPrint"]/div/div/div[3]/ul/li[1]/p'
            year = driver.find_element(By.XPATH, year_id)
            # 기업규모
            size_id ='//*[@id="divContentsPrint"]/div/div/div[3]/ul/li[2]/p'
            size = driver.find_element(By.XPATH, size_id)
            # 근로자수
            employee_id = '//*[@id="divContentsPrint"]/div/div/div[3]/ul/li[4]/p'
            employee = driver.find_element(By.XPATH, employee_id)


            # 페이지 소스 가져오기
            page_source = driver.page_source

            # BeautifulSoup으로 파싱
            soup = BeautifulSoup(page_source, 'html.parser')

            # ID로 요소 찾기
            element = soup.find(id="address1")

            # 텍스트 추출
            address = element["value"]


            # 클래스로 요소 찾기 및 텍스트 추출
            elements = soup.find_all(class_="wel03")
            text1 = [element.get_text(strip=True) for element in elements]

            elements = soup.find_all(class_="wel04 l_bd")
            text2 = [element.get_text(strip=True) for element in elements]

            elements = soup.find_all(class_="wel05")
            text3 = [element.get_text(strip=True) for element in elements]

            elements = soup.find_all(class_="wel06 l_bd")
            text4 = [element.get_text(strip=True) for element in elements]

            # 복지 개수 세기 0 ~ 4
            text_list = [text1, text2, text3, text4]
            value = sum([1 for text in text_list if text])

            # 데이터프레임에 추가
            df.loc[i] = [company, year.text,size.text,employee.text, address,value]

        except NoSuchElementException:
            # 연관 검색어와 기업정보 버튼 모두 찾지 못한 경우에 대한 예외 처리
            print(f"{company}기업정보 버튼과 연관 검색어를 모두 찾지 못했습니다.")

    # 드라이버 종료
    driver.quit()
    
    i += 1 
```
